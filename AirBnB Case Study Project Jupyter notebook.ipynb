```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('files/AB_NYC_2019.csv')
air_bnb.head()
```

```python
# How many neighborhood groups are available and which shows up the most?

# print(air_bnb['neighbourhood_group'].describe())

print(f" There are {air_bnb['neighbourhood_group'].describe()['unique']} neighborhood groups available.")
print(f" \n The {air_bnb['neighbourhood_group'].describe()['top']} neighborhood group appears the most with a count of {air_bnb['neighbourhood_group'].describe()['freq']}.")


# Create a new count column
air_bnb['count'] = 1
print(f" \n Neighborhood Group Summary: \n {air_bnb.groupby(['neighbourhood_group']).count()['count']}")
```

```python
# Are private rooms the most popular in manhattan?

#Same results as below
# room_types = air_bnb.groupby(['neighbourhood_group', 'room_type']).count()
# print(room_types['id'])


air_bnb['count'] = 1
room_types = air_bnb.groupby(['neighbourhood_group', 'room_type']).count()['count'] 
print(f" Room Type Counts by Neighborhood Group: \n {room_types}")
print(f" \n Private rooms are the most popular in the Brooklyn neighborhood group with 10132. \n Private rooms are second most popular in Manhattan with 7982.")

print('\n')

print(f" Looking at only Manhattan, Entire home/apt room types are the most popular at 13199, \n followed by Private rooms at 7982, and Shared rooms at 480. \n")
print(f"\n Manhattan Room Type Summary:")
manhattan = pd.DataFrame(air_bnb[(air_bnb['neighbourhood_group'] == 'Manhattan')].groupby(('room_type')).count()['count'])
manhattan
```

```python
# Which hosts are the busiest and based on their reviews?

# Sum reviews by host_id, then sort by number of reviews
busiest_hosts = air_bnb.groupby(['host_id', 'host_name']).sum()[ ['number_of_reviews']].sort_values(by=['number_of_reviews'],ascending=False) 


# Use .head() to return top 5
print(f"Top 5 Busiest Hosts Based on Number of Reviews: \n")
busiest_hosts.head()


# air_bnb[['host_name','host_id']]
```

```python
# Which neighorhood group has the highest average price?

# air_bnb['avg_price'] = air_bnb.groupby('neighbourhood_group').mean(['price']).sort_values(by=['price'],ascending=False)
# air_bnb

# avg_price = air_bnb.groupby('neighbourhood_group').mean()[ ['price']].sort_values('price',ascending=False)
# air_bnb['average_price_by_neighborhood'] = avg_price
# air_bnb


# air_bnb['avg_price'] = air_bnb.groupby('neighbourhood_group')['price'].transform('avg_price')
# air_bnb


# Manually calculate average
# avg_price = air_bnb.groupby('neighbourhood_group').sum()[['price']] // air_bnb.groupby('neighbourhood_group').count()
# air_bnb


avg_price = air_bnb.groupby('neighbourhood_group').mean()[['price']].sort_values('price',ascending=False)
print(f" Manhattan has the highest average price at $196.88")
avg_price
```

```python
# Which neighbor hood group has the highest total price?


print(f" The Brooklyn, Manhattan, and Queens neighborhood groups are tied with a highest total price of 10000.")

max_price = air_bnb.groupby(['neighbourhood_group']).max(['price']).sort_values(by=['price'],ascending=False)
max_price[['price']]

```

```python
#Which top 5 hosts have the highest total price?

highest_price_host = air_bnb[['host_id','host_name','price']].sort_values('price',ascending=False)

print(f" Top 5 hosts with the highest total price: ")
# .head() returns top 5
highest_price_host.head()
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?

# air_bnb[['host_name','host_id']]


# Results by number of reviews by host id
hosts = air_bnb.groupby('host_id').sum().sort_values('number_of_reviews',ascending=False) 
no_avail = hosts.loc[(hosts['number_of_reviews'] >=100 ) & (hosts['availability_365'] == 0)]
print(f" 136 Results for Zero Availability with a Review Count of 100 or More - by Host ID: \n")
print(no_avail [['number_of_reviews','availability_365']])


# Results by number_of_reviews by property (no deduping)
no_avail = air_bnb[(air_bnb['availability_365']== 0) & (air_bnb['number_of_reviews']>= 100)].sort_values(by=['number_of_reviews'],ascending=False)
print(f" \n 162 Results for Zero Availability with a Review Count of 100 or More - by Property ID: \n")
no_avail [['name','id','host_name', 'host_id', 'availability_365', 'number_of_reviews']]
```

```python
# What host has the highest total of prices and where are they located?

print(f" Sonder (NYC), located in the Manhattan neighborhood group, has the highest total of prices at $82,795:")
# Sum prices by host_id, by location
host_highest_total = air_bnb.groupby(['host_id', 'host_name', 'neighbourhood_group']).sum()[ ['price']].sort_values(by=['price'],ascending=False) 

host_highest_total
```

```python
# When did Danielle from Queens last receive a review?

print(f" Danielle from Queens last received a review on 2019-07-08:")
danielle_queens = air_bnb[(air_bnb['host_name']== 'Danielle') & (air_bnb['neighbourhood_group']== 'Queens')].sort_values('last_review', ascending = False)
# Display five most recent
danielle_queens[['host_name', 'neighbourhood_group', 'last_review']].head()

```
